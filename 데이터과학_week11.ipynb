```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_csv('kuflix.csv')
result = apriori(df, min_support = 0.5, use_colnames=True)
 # min_support: 몸집 작은 놈 적당히 보내야 할 때 그 기준, 수많은 규칙 존재할 수 있겠지만 실제 출력되는 건 지지도가 50%이상인 집합뿐이다. 연관 규칙에 대한 결과는 없고 순수하게 지지도가 0.5 이상인 집합만 나온다.
# 더 보고 싶으면 min_support 값을 줄이면 된다. 0.0은 안된다. 아이템셋 배제
```

```python
result # 자기가 시청한 프로그램 집합에 나솔이 있을 확률이 절반이다.
```

```python
result2 = association_rules(result, metric='confidence', min_threshold=0.5) #신뢰도 1/2도 안되면 규칙이냐? metric 신뢰도로 잡아서 신뢰도가 0.5도 안되면 규칙으로 출력도 하지 마세요

#되도록 많은 규칙 살아남도록 하려면 0.0에 가깝게 두기
```

```python
result2 #나솔 본 모든 사람이 최강 야구 봤대(1)~confidence 보기, antecedents가 앞조건, consequents가 뒷조건
```
